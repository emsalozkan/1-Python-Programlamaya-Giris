```python
l=[2,13,18,93,22]

def func(list):
  TEK=[]
  CIFT=[]

  for indx in range(len(list)):
    if list[indx]%2 == 0:
      CIFT.append(list[indx])
    else:
      TEK.append(list[indx])

  return CIFT,TEK

even_list,odd_list = func(l)
print(even_list)
print(odd_list)
```

```python
l=[2,13,18,93,22]

def func(list):
  TEK=[]
  CIFT=[]

  for i in list:
    if i%2 == 0:
      CIFT.append(i)
    else:
      TEK.append(i)

  return CIFT,TEK

even_list,odd_list = func(l)
print(even_list)
print(odd_list)
```
