```python
#15 survived değişkeninin pclass ve sex değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.

df.groupby(["pclass", "sex"]).agg({"survived": ["sum","count","mean"]}) #sum hayatta kalanları veriyor
```

```python
#16 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 verecek bir fonksiyon yazın. 
#Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz. 
#(apply ve lambda yapılarını kullanınız)
```

```python
df["age_flag"] = df["age"].apply(lambda x: 1 if x<30 else 0)
df
```

```python
# 17 Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
# 18 Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill değerinin sum, min, max ve mean değerlerini bulunuz.

df.groupby(["time"]).agg({"total_bill": ["sum","min","max","mean"]}) 
```

```python
# 19 Day ve time’a göre total_bill değerlerinin sum, min, max ve mean değerlerini bulunuz.

df.groupby(["day","time"]).agg({"total_bill": ["sum","min","max","mean"]}) 
```

```python
#20 Lunch zamanına ve kadın müşterilere ait total_bill ve tip  değerlerinin day'e göre sum, min, max ve mean değerlerini bulunuz

df.loc[(df["sex"] == "Female") & (df["time"] == "Lunch")].groupby(["day"]).agg({"total_bill": ["sum","min","max","mean"],
                                                                                "tip": ["sum","min","max","mean"]}) 
```

```python
# 21 size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir? (loc kullanınız)

df.loc[(df["size"] < 3) & (df["total_bill"] > 10), ["total_bill"]].mean()
```

```python
#22 total_bill_tip_sum adında yeni bir değişken oluşturunuz. Her bir müşterinin ödediği totalbill ve tip in toplamını versin.

df["total_bill_tip_sum"] = df["total_bill"]+df["tip"]
df
```

```python
# 23 total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.

new_df= df.sort_values ("total_bill_tip_sum", ascending=False)[:30]
new_df.shape
```
